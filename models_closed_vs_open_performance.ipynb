```python
def licence_to_color(licence):
    if licence=="Closed":
        return "red"
    elif licence=="Non commercial":
        return "orange"
    elif licence=="Open weights":
        return "green"
    elif licence=="Open source":
        return "lime"
    else:
        return "black"
```

```python
def model_name_to_marker(name):
    name = name.lower()
    if name.startswith("gpt-3.5"):
        return '*'
    elif name.startswith("gpt-4"):
        return 'X'
    elif name.startswith("claude"):
        return 'p'
    elif name.startswith("llama"):
        return 's'
    elif name.startswith("mi"):
        return 'P'
    elif name.startswith("qwen"):
        return '^'
    else:
        return '.'
```
